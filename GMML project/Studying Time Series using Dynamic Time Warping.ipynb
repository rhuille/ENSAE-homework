Voici l'implementation de la dynamic time warping distance : 

```python
def DTWDistance(s,t):
    """ 
    Dynamic time warping
    --------------------
    return the DTW between the numpy array s and t with euclidian distance as ground distance
    """
    n,m = s.size, t.size
    DTW = np.full((n,m), np.infty)
    DTW[0, 0] = 0

    for i in range(1,n):
        for j in range(1,m):
            cost = (s[i]- t[j])**2
            DTW[i, j] = cost + np.min([DTW[i-1,j],DTW[i,j-1], DTW[i-1,j-1]])

    return np.sqrt(DTW[n-1,m-1])
```

On la teste sur un exemple : 

```python
s = np.array([0,0,6,0,0,2,0,0])
t = np.array([0,0,0,5,0,3,0,0])

plt.plot(s)
plt.plot(t)
plt.show()

print('DTWDistance (perso) : ', DTWDistance(s,t))
print('DTWDistance (tslearn) : ', dtw(s,t))
print('Euclidian Distance : ', np.sum((s-t)**2) )
```

Elle fonctionne aussi sur des séries de longueurs différentes : 

```python
s = np.array([0,0,6,0,0,2,0,0])
t = np.array([0,0,0,5,0,0,0,3,0,0,1,2,5,0,0])

plt.plot(s)
plt.plot(t)
plt.show()

print('DTWDistance (perso) : ', DTWDistance(s,t))
print('DTWDistance (tslearn) : ', dtw(s,t))
```

#### Knn with my own implementation

```python
def knn(i_test,y,DistanceMatrix, k = 3):
    """
    K Nearest Neighbors
    -------------------
    
    return the predicted class for the test set givin by the index 'i_test' 
    """
    
    n = DTWMatrix.shape[0]
    y_predict = []
    i_train = np.setdiff1d(np.arange(n), i_test)
    
    for i in i_test : 
        l = DistanceMatrix[i,:].copy()
        l[i_test] = np.infty
        a = []
        for r in range(k):
            j = np.argmin(l)
            a+=[j]
            l[j] = np.infty
            
        y_predict += [ np.argmax([sum(y[a] == i) for i in range(1,6)])+1 ]
    
    return np.array(y_predict)
```

On obtient : 

```python
i_index = np.random.choice(np.arange(150), 10)
y_test = y[i_index]
y_predict = knn(i_index, y, DTWMatrix)
print('Accuracy score : ', sum(y_predict == y_test)/y_test.size )
```

### Question 3

*In order to visualize representative examples for each class in these samples, we consider time series averaging. We will compute an average for each class in the 300 examples selected for the previous question. Code to compute such averages is available in R or Python. Plot these averages and compare them with the usual Euclidean averaging you would obtain by simply summing these time-series.*

```python
from tslearn.barycenters import dtw_barycenter_averaging
```

```python
DBA = []
for i in tqdm(range(1,6)):
    DBA += [dtw_barycenter_averaging(X[y==i])]
```

```python
plt.figure(figsize=(15,10))
plt.title('Averaging with DBA')
for t in DBA : 
    plt.plot(t)
plt.show()
```

```python
Euclidian = []
for i in tqdm(range(1,6)):
    Euclidian += [X[y==i].mean(0)]
```

```python
plt.figure(figsize=(15,10))
plt.title('Averaging with Euclidian distance')
for t in Euclidian : 
    plt.plot(t)
plt.show()
```
