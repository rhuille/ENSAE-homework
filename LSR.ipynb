# SGD

```python
n = 100
p = 5

X = pi*np.dot(np.reshape(np.arange(1,n+1), (n,1)), 
              np.reshape(np.arange(1,p+1), (1,p)))/n
X = np.cos(X)
theta_ = np.reshape(np.arange(1,p+1), (p,1))

f = np.dot(X,theta_)

e = np.reshape(np.random.normal(size=n, scale=2), (n,1))
y = f + e


plt.figure(figsize=(10,5))
plt.plot(np.arange(1,n+1)/n, f, 'b--', 
         np.arange(1,n+1)/n, y, 'k.')
plt.show()
```

```python
# descente de gradient normal

# parameters : 
theta = np.full((p , 1), 0) # init
eta = 0.001  # step

list_theta = np.array(theta)
list_norm_grad = []
for i in range(100):
    # computation of the gradient : 
    grad = 2*np.dot(X.T, np.dot(X, theta) - y)
    # next value of theta
    theta = theta - eta*grad
    list_theta = np.hstack([list_theta, theta])
    list_norm_grad += [np.linalg.norm(grad)]

plt.figure(figsize=(10,4))
plt.plot(list_norm_grad, '-')
plt.show()

plt.figure(figsize=(10,4))
plt.plot(list_theta[0,:], list_theta[1,:].T, 'ko-')
plt.show()

plt.figure(figsize=(10,4))
plt.plot(list_theta.T, '-')
plt.show()
```
