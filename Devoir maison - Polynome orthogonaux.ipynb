# Polynôme de Tchebychev

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Tchebytchev = [] # liste contenenant les polynomes de techbytchev
# Un polynome de degres n est representé par le numpy array de ses n coeeficients

Tchebytchev = [
    np.array(0),
    np.array([1, 0]),
    np.array([2, 0, -1]),
]
```

```python
# 2eme polynome de Tchebytchev:
Tchebytchev[2]
```

```python
def polynome_function(x, coeffs):
    y = 0
    n = len(coeffs)
    for i in range(n):
        y += coeffs[i]*x**(n-i-1)
    return y
```

```python
x = np.linspace(-1,1,100)
plt.plot(x, polynome_function(x, Tchebytchev[2]))
plt.show()
```

```python
def compute_Tchebytchev(n):
    # initioalisation:
    Tchebytchev = [
        np.array([0]),
        np.array([1, 0]),
        np.array([2, 0, -1]),
    ]
    
    for i in range(1, n+2):
        Tchebytchev.append(2*np.append(Tchebytchev[i+1], 0) - np.append([0,0], Tchebytchev[i]))
    
    return Tchebytchev
```

```python
Tchebytchev = compute_Tchebytchev(12)
```

```python
x = np.linspace(-1,1,100)
plt.plot(
    x, polynome_function(x, Tchebytchev[1]), 
    x, polynome_function(x, Tchebytchev[2]), 
    x, polynome_function(x, Tchebytchev[7]),
    x, polynome_function(x, Tchebytchev[12])
)
plt.show()
```
